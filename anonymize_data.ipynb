```python

df = df.sort_values(by='Last Name')
first_150 = df.head(150)
```

```python
first_150.info()
```

```python

import random
import string

# define a function to generate random strings
def random_string(length):
    return ''.join(random.choice(string.ascii_letters) for i in range(length))

# define the list of columns to anonymize
columns_to_anonymize = ['Last Name', 'First Name']

# anonymize the columns with random values
for column in columns_to_anonymize:
    first_150[column] = [random_string(10) for i in range(len(first_150))]

# anonymize the Daytime Phone and Email Address columns
for index, row in first_150.iterrows():
    first_150.at[index, 'Daytime Phone'] = f"({random.randint(100, 999)}){random.randint(100, 999)}-{random.randint(1000, 9999)}"
    first_150.at[index, 'Email Address'] = f"{random_string(5)}@{random_string(5)}.com"
    
# anonymize the Advisor Email column
first_150['Advisor Email'] = [random_string(8) + '@gmail.com' for i in range(len(first_150))]
```
